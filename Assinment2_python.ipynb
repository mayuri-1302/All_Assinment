```python
# Q4.Create a simple calculator using if-else conditions.
x = int(input("Enter a number:"))
y = int(input("Enter a number:"))
Choice = input("Enter an operator (+, -, *, /):")
if Choice == "+":
    print(x, "+", y, "=", x + y)
elif Choice == "-":
    print(x, "-", y, "=", x - y)
elif Choice == "*":
    print(x, "*", y, "=", x * y)
elif Choice == "/":
    if y != 0:
        print(x, "/", y, "=", x / y)
    else:
        print("Error: Division by zero is not allowed.")
```

```python
# Q7.Write a program to find the factorial of a number.
num = 10
fact = 1
i = 1
while i <= num:
    fact = fact * i
    i = i + 1
    print(fact)
```

```python
# Q8.Reverse a given number using loops.
number = int(input("enter a number:"))
reverse = 0
while number > 0:
    digit = number % 10
    reverse = reverse * 10 + digit
    number = number // 10
print("The reversed number is:",reverse)
```

```python
# Q9.Print the following pattern
# *
# **
# ***
# ****
# *****
rows = 5
for i in range (1,rows+1):
    stars = "*"*i
    print(stars)
```

```python
# Q10. Print the following pattern
#    *
#   ***
#  *****
# *******
rows = 5

for i in range(1, rows + 1):
    spaces = " " * (rows - i)
    stars = "* " * i
    print(spaces + stars)
```
