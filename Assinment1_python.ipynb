Q3. Convert:

Integer → Float

Float → Integer

Integer → String

Print before and after conversion.

```python
# Integer to Float
num_int = 10
print("Before Conversion:", num_int, type(num_int))

num_float = float(num_int)
print("After Conversion to Float:", num_float, type(num_float))

# Float to Integer
float_num = 15.1
print("\nBefore Conversion:", float_num, type(float_num))

int_num = int(float_num)
print("After Conversion to Integer:", int_num, type(int_num))

# Integer to String
integer_num = 25
print("\nBefore Conversion:", integer_num, type(integer_num))

string_num = str(integer_num)
print("After Conversion to String:", string_num, type(string_num))
```

Q4. Create a list of 5 fruits:

Print first item
Print last item
Add one fruit
Remove one fruit
Reverse list

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]
print(fruits)
print("First Item =", fruits[0])

print("Last Item =", fruits[len(fruits)-1])

fruits.insert(2, "Pineapple")
print("After Adding =", fruits)

del fruits[1]
print("After Removing =", fruits)

reversed_list = fruits[::-1]
print("Reversed List =", reversed_list)
```

Q5. Given marks=[70,80,90,60,100]
Find:

Maximum
Minimum
Average

```python
marks = [70, 80, 90, 60, 100]

print("Maximum =", max(marks))

print("Minimum =", min(marks))

Average = sum(marks) / len(marks)
print("Average =", Average)
```
